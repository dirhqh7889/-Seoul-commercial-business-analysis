# 구별 학원수 비교 분석
    *위에서 분석해본 결과 강남구에 상가업소가 가장 많은 것으로 알게 되었다.
    *그 중 학원수를 분석하여 과연 학원 교육의 정점으로 알려진 대치동에 많은 학원이 분포하는지 확인할 것이다.

```python
#"시도명"이 "서울특별시"이고 "상권업종대분류명"이 "학문/교육"인 데이터를 서브셋으로 가져온다
# 재사용을 위해 서브셋을 df_academy에 저장
df_academy = df[(df["시도명"]=="서울특별시") & (df["상권업종대분류명"] == "학문/교육")].copy()
df_academy
```

```python
academy_count = df_academy["상권업종소분류명"].value_counts().head(30)
```

```python
academy_count_a = academy_count[academy_count > 4000]
academy_count_a
```

```python
df_academy_select = df_academy[df_academy["상권업종소분류명"].isin(academy_count_a.index)].copy()
df_academy_select
```

# folium으로 지도 보여주기

```python
import folium
```

```python
long = df_academy_select["경도"].mean()
lat = df_academy_select["위도"].mean()
```

```python
df_m = df_academy_select[df_academy_select["상권업종소분류명"].isin(["학원-입시"])]
df_m = df_m.sample(1000)
```

```python
df_m.iloc[0]
```

```python
m = folium.Map(location=[lat, long], zoom_start = 11.5, tiles="Stamen Terrain")

for i in df_m.index[:100]:
    tooltip = df_m.loc[i,"상호명"] + "-" + df_m.loc[i,"도로명주소"]
    lat = df_m.loc[i,"위도"]
    long = df_m.loc[i,"경도"]
    folium.Marker([lat,long],  tooltip="tooltip", radius = 3).add_to(m)
m
```

```python
m.save('mapindex.html')
```

```python
from folium.plugins import MarkerCluster
m = folium.Map([lat, long], zoom_start = 11)
marker_cluster = MarkerCluster().add_to(m)

for i in df_m.index:
    sub_lat = df_m.loc[i,"위도"]
    sub_long = df_m.loc[i,"경도"]
    title = df_m.loc[i,"상호명"] + "-" +df_m.loc[i,"도로명주소"]
    icon_color = "red"
    folium.Marker(
        [sub_lat, sub_long],
        icon=folium.Icon(color=icon_color),
        popup=title,
        tooltip=title).add_to(marker_cluster)
    
m
```
